```python
print("\nHOST INFORMATION")
print("-"*70)

hostname = socket.gethostname()

print("Hostname: ", hostname)

try:
  ip_address = socket.gethostbyname(hostname)
  print("IP Address: ", ip_address)
except :
  print("Unable to Retrieve IP Address")
```

```python
print("\nINTERNET CONNECTIVITY TEST")
print("-"*70)

try:
  response = requests.get("https://github.com/PANARA195/3040233402_Information-Security/blob/main/IS_PRACTICAL_1.pdf")

  print("Connection Status : SUCCESS")
  print("HTTP Status Code : ", response.status_code)

except:
  print("Connection Status : Failed")
```

```python
print("\nDNS RESOLUTION TEST")
print("-"*70)

try:
  github_ip - socket.gethostbyname("https://github.com/PANARA195/3040233402_Information-Security/blob/main/IS_PRACTICAL_1.pdf")

  print("Domain Name : github.com")
  print("Resolved IP :", github_ip)

except:
  print("DNS Resolution Failed")
```

```python
print("\nPORT CONNECTIVITY TEST")
print("-"*70)

host = "google.com"

ports = [80, 443]

for port in ports:
  try:
    connection = socket.create_connection((host,port), timeout=5)
    print(f"Port {port} is open")
    connection.close()
  except:
    print(f"Port {port} is closed")
```
